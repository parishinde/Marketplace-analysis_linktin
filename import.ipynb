```python
import pandas as pd
import sqlite3

# CSV file path (same folder)
csv_file = "postings_utf8.csv"

# SQLite database file
db_file = "jobs.db"

# Read CSV
df = pd.read_csv(csv_file)

# Connect to SQLite
conn = sqlite3.connect(db_file)

# Insert data into table
df.to_sql("rawww", conn, if_exists="append", index=False)

# Close connection
conn.close()

print("CSV imported successfully into SQLite!")
```
